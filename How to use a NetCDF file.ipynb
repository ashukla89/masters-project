```python
nc_fid = Dataset("prescomp.GVYOub4nL3.nc", "r")

lons = nc_fid.variables['lon'][:]
lats = nc_fid.variables['lat'][:]
sst = nc_fid.variables['sst'][:]
sst_units = nc_fid.variables['sst'].units

nc_fid.close()

import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np

# Get some parameters for the Stereographic Projection
lon_0 = lons.mean()
lat_0 = lats.mean()

m = Basemap(width=12000000,height=9000000,
            resolution='l',projection='eqdc',\
            lat_1=45.,lat_2=55,lat_0=50,lon_0=-180.)

lon, lat = np.meshgrid(lons, lats)
xi, yi = m(lon, lat)

# Plot Data
contour = m.contourf(xi,yi,np.squeeze(sst))
print("Drawing contour at", contour.levels)

m.contour(xi,yi,np.squeeze(sst), levels=contour.levels)
print("Drawing contour lines ", contour.levels)

# Add Grid Lines
m.drawparallels(np.arange(-80., 81., 10.), labels=[1,0,0,0], fontsize=10)
m.drawmeridians(np.arange(-180., 181., 10.), labels=[0,0,0,1], fontsize=10)

# Add Coastlines, States, and Country Boundaries
m.drawcoastlines()
m.drawstates()
m.drawcountries()
```

```python
import matplotlib.pyplot as plt

# CHANGE
# Set the figsize, because you can't do it with basemap
plt.figure(figsize=(16,6))

nc_fid = Dataset("prescomp.GVYOub4nL3.nc", "r")

lons = nc_fid.variables['lon'][:]
lats = nc_fid.variables['lat'][:]
sst = nc_fid.variables['sst'][:]
sst_units = nc_fid.variables['sst'].units

nc_fid.close()

import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np

# Get some parameters for the Stereographic Projection
lon_0 = lons.mean()
lat_0 = lats.mean()

m = Basemap(width=12000000,height=9000000,
            resolution='i',projection='eqdc',\
            lat_1=45.,lat_2=55,lat_0=50,lon_0=-180.)

lon, lat = np.meshgrid(lons, lats)
xi, yi = m(lon, lat)



# CHANGE
# Pick our own levels
levels = [-4, -2, 0, 2, 4]
# Or be lazy about it, this is -5 to 5 in steps of 0.5
levels = np.arange(-5, 5, 0.75)
# Or just make it automatic
levels = None

# Fill in contour areas
contour = m.contourf(xi,yi,np.squeeze(sst), levels=levels)
print("Drawing contour at", contour.levels)

# Draw lines on top of it based on the same areas
# by default they're the same colors, but we want
# them to be black instead (k)
lines = m.contour(xi,yi,np.squeeze(sst),
          levels=contour.levels,
          colors='k')
print("Drawing contour lines ", contour.levels)

# And label the contour lines, too, because why not
plt.clabel(lines, inline=1, fontsize=10)




# Add Grid Lines
m.drawparallels(np.arange(-80., 81., 10.), labels=[1,0,0,0], fontsize=10)
m.drawmeridians(np.arange(-180., 181., 10.), labels=[0,0,0,1], fontsize=10)

# Add Coastlines, States, and Country Boundaries
m.drawcoastlines()
m.drawstates()
m.drawcountries()
```

```python
import numpy as np

filenames = glob.glob("*.nc")
for filename in filenames:
    print("Processing", filename, "into csv")

    rows = []
    for i, lat in enumerate(lats):
        for j, lon in enumerate(lons):
            if sst[0][i][j] != '--':
                val = sst[0][i][j]
            else:
                val = np.nan
            datapoint = {
                'lat': lat,
                'lon': lon,
                'sst': val
            }

            rows.append(datapoint)
    csv_filename = filename.replace(".nc", ".csv")
    pd.DataFrame(rows).to_csv(csv_filename, index=False)
    print("\tSaved as", csv_filename)
```
