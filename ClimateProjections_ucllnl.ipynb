```python
# first, know that you have to change from mm into in
mm_to_in = 0.0393701

# also, that Centrigrade needs to be rendered as Fahrenheit
# we'll write a function to do this conversion
def c_to_f(t):
    t = t*9/5 + 32
    return t
```

Now let's do some analysis just on precipitation to see if this makes any sense:

```python
# aggregate by monthly averages (first sum into months, then average the month across years)

print('Average precip of all years-- RCP4.5:')
print(pr.groupby(['year','month']).avg_rcp45.sum().reset_index().groupby('month').avg_rcp45.mean() * mm_to_in)
print(pr.groupby(['year','month']).avg_rcp45.sum().reset_index().groupby('month').avg_rcp45.mean().sum() * mm_to_in)

print('Average precip of all years-- RCP8.5:')
print(pr.groupby(['year','month']).avg_rcp85.sum().reset_index().groupby('month').avg_rcp85.mean() * mm_to_in)
print(pr.groupby(['year','month']).avg_rcp85.sum().reset_index().groupby('month').avg_rcp85.mean().sum() * mm_to_in)

# the below results seem a bit high. the avg annual precip through 2010 is 29.42 inches per Wikipedia: https://en.wikipedia.org/wiki/Sonoma,_California

# let's try just the historical period
print('Average precip of historical years:')
print(pr[pr['year']<=1999].groupby(['year','month']).avg_rcp45.sum().reset_index().groupby('month').avg_rcp45.mean().sum() * mm_to_in)
print(pr[pr['year']<=1999].groupby(['year','month']).avg_rcp85.sum().reset_index().groupby('month').avg_rcp85.mean().sum() * mm_to_in)

# and now future period
print('Average precip of future years:')
print(pr[pr['year']>2010].groupby(['year','month']).avg_rcp45.sum().reset_index().groupby('month').avg_rcp45.mean().sum() * mm_to_in)
print(pr[pr['year']>2010].groupby(['year','month']).avg_rcp85.sum().reset_index().groupby('month').avg_rcp85.mean().sum() * mm_to_in)
```

The above mostly makes sense, although precipitation seems a bit high. Perhaps the data reflects a hillside location with higher precip.

Let's go ahead and write a function that will do all the above work for other prediction datasets we might encounter in this format (taking into account the fact that some will be precipitation, and others temperature predictions).

```python
# first we need a helper function that figures out what decade something is in
# as that will be key to our analysis
def decade(year):
    decade_start = int(np.floor(year/10)*10)
    decade_end = decade_start + 9
    decade = str(decade_start) + "-" + str(decade_end)
    return decade
```

```python
# the function takes the df to feed in, and a string that is the name of the measure we care about
def transform_proj(df,measure_name):
    # change column names
    df.columns = columns
    # generate an RCP4.5 average
    df[measure_name+'_rcp45'] = df[rcp45].mean(axis=1)
    # generate an RCP8.5 average
    df[measure_name+'_rcp85'] = df[rcp85].mean(axis=1)
    # filter to just that average
    df = df[['year','month','date',measure_name+'_rcp45',measure_name+'_rcp85']]
    # add column for decade, for later analysis
    df['decade'] = df['year'].apply(decade)
    # put decade column in front
    # here's a handy trick for that: https://stackoverflow.com/questions/25122099/move-column-by-name-to-front-of-table-in-pandas
    df = df[ ['decade'] + [ col for col in df.columns if col != 'decade' ] ]
    
    # transform the measures
    # from mm to inches if it's precip
    if measure_name == 'pr':
        df[measure_name+'_rcp45'] = df[measure_name+'_rcp45'] * mm_to_in
        df[measure_name+'_rcp85'] = df[measure_name+'_rcp85'] * mm_to_in
    # from C to F if it's temp
    if 'tas' in measure_name:
        df[measure_name+'_rcp45'] = c_to_f(df[measure_name+'_rcp45'])
        df[measure_name+'_rcp85'] = c_to_f(df[measure_name+'_rcp85'])
        
    return df
```

```python
# let's do the same for tmax and tmin

# read in the max temperature with no headers
tasmax = pd.read_csv('tasmax_son.csv', header=None)
tasmax = transform_proj(tasmax,"tasmax")

# read in the min temp with no headers
tasmin = pd.read_csv('tasmin_son.csv', header=None)
tasmin = transform_proj(tasmin,"tasmin")

# also, re-read pr for consistency
pr = pd.read_csv('pr_son.csv', header=None)
pr = transform_proj(pr,"pr")
```

### Let's also pull and process historical data for comparison

```python
# read in each file, but just for historicals

prh = pd.read_csv('pr_son_h.csv', header=None)
tasmaxh = pd.read_csv('tasmax_son_h.csv', header=None)
tasminh = pd.read_csv('tasmin_son_h.csv', header=None)
```

```python
# write a function that transforms each of these types of dfs

# measure_name should only be 'pr' or 'tasmax' or 'tasmin' !!
def transform_hist(df,measure_name):
    # add column names (no climate projections here)
    # list the columns
    cols_h = ['year','month','date',measure_name]
    # add the names
    df.columns = cols_h
    # add column for decade, for later analysis
    # basically, figure out the start of the decade and the end, and turn them into a string
    df['decade'] = df['year'].apply(decade)
    # put decade column in front
    # here's a handy trick for that: https://stackoverflow.com/questions/25122099/move-column-by-name-to-front-of-table-in-pandas
    df = df[ ['decade'] + [ col for col in df.columns if col != 'decade' ] ]
    
    # replace missing values when represented by '--'
    df.replace('--',0,inplace=True)
    # convert all columns to floats
    df.iloc[:,4:] = df.iloc[:,4:].astype(float)

    # transform the measures
    # from mm to inches if it's precip
    if measure_name == 'pr':
        df[measure_name] = df[measure_name] * mm_to_in
    # from C to F if it's temp
    if 'tas' in measure_name:
        df[measure_name] = c_to_f(df[measure_name])
        
    # rename the column to reflect that it's historical
    df.rename({measure_name:measure_name+'_hist'},axis='columns',inplace=True)
    
    # return the df
    return df
```

```python
prh = transform_hist(prh,'pr')
tasmaxh = transform_hist(tasmaxh,'tasmax')
tasminh = transform_hist(tasminh,'tasmin')
```

### Let's combine the dataframes into one for easier analysis

```python
# specify the columns to do the merge on
on_cols = ['year','month','date','decade']

# merge all the dfs into 1
# define a function to do that

def combine_dfs(df_pr_hist,df_tasmax_hist,df_tasmin_hist,df_pr_proj,df_tasmax_proj,df_tasmin_proj):
    df_combo = df_pr_hist.merge(df_tasmax_hist,how='outer',left_on=on_cols,right_on=on_cols).\
        merge(df_tasmin_hist,how='outer',left_on=on_cols,right_on=on_cols).\
        merge(df_pr_proj,how='outer',left_on=on_cols,right_on=on_cols).\
        merge(df_tasmax_proj,how='outer',left_on=on_cols,right_on=on_cols).\
        merge(df_tasmin_proj,how='outer',left_on=on_cols,right_on=on_cols)
    
    # reorder the columns as desired
    # specifically, just moving the 'decade' column up to the front
    # here's a handy trick for that: https://stackoverflow.com/questions/25122099/move-column-by-name-to-front-of-table-in-pandas
    df_combo = df_combo[ ['decade'] + [ col for col in df_combo.columns if col != 'decade' ] ]
    
    return df_combo

combo = combine_dfs(prh,tasmaxh,tasminh,pr,tasmax,tasmin)
```

For ease of comparison, let's build a summary side-by-side table comparing historicals and predictions. We can just use RCP4.5 for this, since it's almost the same as RCP8.5.

```python
# build a function to do that

def historical_comparison(df):
    # Recall that temperatures are just a monthly average
    df_comp_pr = df[df['year']<=1999].groupby(['year','month'])['pr_hist','pr_rcp45'].sum().reset_index().groupby('month')['pr_hist','pr_rcp45'].mean()
    # While precipitation is an average of the sum for a whole calendar month
    df_comp_t = df[df['year']<=1999].groupby('month')['tasmax_hist','tasmax_rcp45',\
                                                         'tasmin_hist','tasmin_rcp45'].mean()
    df_comp = df_comp_pr.merge(df_comp_t,left_index=True,right_index=True)
    return df_comp

historical_comparison(combo)
```

They're pretty close-- but the temperature predictions have a systematically lower bias (by over a degree in summer nights) compared to the historicals.
```python
#combo.groupby(['decade','month'])[['tasmax_rcp85']].mean()

def illustrate_range(df):

    # determine min-max range of each measure, so we know where to set the ylims on the graph
    # the double-max/ double-min is because the max just gives us the max/min for each column
    # we want the max/ min across BOTH columns
    tasmax_max = df.groupby(['decade','month'])[['tasmax_rcp45','tasmax_rcp85']].mean().max().max() + 5
    tasmax_min = df.groupby(['decade','month'])[['tasmax_rcp45','tasmax_rcp85']].mean().min().min() - 5
    
    tasmin_max = df.groupby(['decade','month'])[['tasmin_rcp45','tasmin_rcp85']].mean().max().max() + 5
    tasmin_min = df.groupby(['decade','month'])[['tasmin_rcp45','tasmin_rcp85']].mean().min().min() - 5
    
    pr_max = (df.groupby(['decade','month'])[['pr_rcp45','pr_rcp85']].sum().max().max())/10 + 1
    # we know that some months are rainless, so pr_min will always be 0
    
    # set up 3x2 plot
    fig, axes = plt.subplots(3,2,figsize=(10,15))
    
    # plot max temp scenarios
    df.pivot_table(index='month',columns='decade',values='tasmax_rcp45',aggfunc='mean').\
        plot(ax=axes[0,0],legend=False,ylim=(tasmax_min,tasmax_max),title=('Tasmax - RCP4.5'))
    df.pivot_table(index='month',columns='decade',values='tasmax_rcp85',aggfunc='mean').\
        plot(ax=axes[0,1],legend=False,ylim=(tasmax_min,tasmax_max),title=('Tasmax - RCP8.5'))
    
    # plot min temp scenarios
    df.pivot_table(index='month',columns='decade',values='tasmin_rcp45',aggfunc='mean').\
        plot(ax=axes[1,0],legend=False,ylim=(tasmin_min,tasmin_max),title=('Tasmin - RCP4.5'))
    df.pivot_table(index='month',columns='decade',values='tasmin_rcp85',aggfunc='mean').\
        plot(ax=axes[1,1],legend=False,ylim=(tasmin_min,tasmin_max),title=('Tasmin - RCP8.5'))
    
    # plot precip scenarios
    (df.pivot_table(index='month',columns='decade',values='pr_rcp45',aggfunc='sum')/10).\
        plot(ax=axes[2,0],ylim=(0,pr_max),title=('Precip - RCP4.5'))
    (df.pivot_table(index='month',columns='decade',values='pr_rcp85',aggfunc='sum')/10).\
        plot(ax=axes[2,1],ylim=(0,pr_max),title=('Precip - RCP8.5'))

    plt.show()

illustrate_range(combo)
```

### Number of 104+ days per year

We have a problem if we want to calculate these-- we averaged a bunch of model runs together. We don't actually see extreme temperatues like this left.

To see why, look below.

```python
# output is a graph of this number for each scenario by year, and by decade
# need to develop series

# here's the problem-- we averaged model results together so few extremes are left

# test for 104+
combo_104 = combo.copy()
combo_104['plus104_hist'] = (combo['tasmax_hist']>=104)
combo_104['plus104_rcp45'] = (combo['tasmax_rcp45']>=104)
combo_104['plus104_rcp85'] = (combo['tasmax_rcp45']>=104)

combo_104.groupby('year')['plus104_hist','plus104_rcp45','plus104_rcp85'].sum()
```

So, instead, we have to go back and develop series that DON'T average the runs together, and to look at the runs individually.

So first, we need a version of our processing function for projections that doesn't take an average, and then compare model runs in terms of both averages and extremes.

```python
def transform_proj_noavg(df,measure_name):
    # change column names
    df.columns = columns
    # add column for decade, for later analysis
    df['decade'] = df['year'].apply(decade)
    
    # move decade up to front
    df = df[ ['decade'] + [ col for col in df.columns if col != 'decade' ] ]
    
    # replace missing values when represented by '--'
    df.replace('--',0,inplace=True)
    # convert all columns to floats
    df.iloc[:,4:] = df.iloc[:,4:].astype(float)
    
    # transform the measures
    # from mm to inches if it's precip
    if measure_name == 'pr':
        df.iloc[:,4:] = df.iloc[:,4:] * mm_to_in
    # from C to F if it's temp
    if 'tas' in measure_name:
        df.iloc[:,4:] = (df.iloc[:,4:]).apply(c_to_f)
        
    return df
```

```python
tasmax_all = pd.read_csv('tasmax_son.csv',header=None)
tasmax_all = transform_proj_noavg(tasmax_all,'tasmax')
```

```python
# which scenarios exhibit the most extreme behavior? are there biases?
# check for one year
# split out by rcp4.5 and rcp8.5

def draw_annual_extremes(df,year):
    
    fig, axes = plt.subplots(1,2,figsize=(10,8))

    df[(df['year']==year)].groupby(['month','date','year'])[rcp45].sum().plot(ax=axes[0],legend=False,ylim=(20,130))
    df[(df['year']==year)].groupby(['month','date','year'])[rcp85].sum().plot(ax=axes[1],legend=False,ylim=(20,130))
    
    plt.show()
    
draw_annual_extremes(tasmax_all,2099)
```

```python
# let's define a function that rolls all of that up into a quick test

# assume you have a transformed dataframe already
def plus_104(df_tasmax):
    # define the condition as a boolean
    test_104 = df_tasmax.iloc[:,4:] >= 104
    # merge back into a df with time period
    tasmax_104 = df_tasmax[['decade','year','month','date']].merge(test_104,left_index=True,right_index=True)
    # average these again, and see on average which scenarios give us more such days
    # groupby decade, separating rcp45 and rcp85, and rename the series to reflect this
    # put them back together with pd.concat
    summary_104 = pd.concat([(tasmax_104.groupby('decade')[rcp45].sum().transpose()/10).mean().rename('mean_rcp45'),\
         (tasmax_104.groupby('decade')[rcp85].sum().transpose()/10).mean().rename('mean_rpc85'),\
        (tasmax_104.groupby('decade')[rcp45].sum().transpose()/10).std().rename('stdev_rcp45'),\
         (tasmax_104.groupby('decade')[rcp85].sum().transpose()/10).std().rename('stdev_rpc85')],axis=1)
    return summary_104

def plus_104_harvest(df_tasmax):
    # narrow to just months of Aug, Sep, Oct
    df_tasmax = df_tasmax[df_tasmax['month'].isin([8,9,10])]
    # define the condition as a boolean
    test_104 = df_tasmax.iloc[:,4:] >= 104
    # merge back into a df with time period
    tasmax_104 = df_tasmax[['decade','year','month','date']].merge(test_104,left_index=True,right_index=True)
    # average these again, and see on average which scenarios give us more such days
    # groupby decade, separating rcp45 and rcp85, and rename the series to reflect this
    # put them back together with pd.concat
    summary_104 = pd.concat([(tasmax_104.groupby('decade')[rcp45].sum().transpose()/10).mean().rename('mean_rcp45'),\
         (tasmax_104.groupby('decade')[rcp85].sum().transpose()/10).mean().rename('mean_rpc85'),\
        (tasmax_104.groupby('decade')[rcp45].sum().transpose()/10).std().rename('stdev_rcp45'),\
         (tasmax_104.groupby('decade')[rcp85].sum().transpose()/10).std().rename('stdev_rpc85')],axis=1)
    return summary_104
```
